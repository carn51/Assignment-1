# Load Data

```python
## Load first chosen dataset into a Pandas (or PySpark/Dask) DataFrame

data = pd.read_csv('/content/listings.csv.gz')
df = pd.DataFrame(data)
df.head(3)
```

# Clean Data

```python
df1 = df[['host_since', 'price', 'beds', 'accommodates', 'reviews_per_month', 'review_scores_value', 'instant_bookable']]    #host name for reference
```

```python
df1['price'] = df1['price'].str.replace('$', '').str.replace(',', '').astype(float)

from datetime import datetime

current = datetime.now()
df1['host_since'] = pd.to_datetime(df1['host_since'])
df1['host_exp'] = ((current - df1['host_since']) / pd.Timedelta(days= 365.25))

df1['host_exp'].head(3)
df2 = df1.drop(columns = ['host_since']).dropna()
```

# Regression Problem

Outliers

```python
iqr = np.quantile(df2['price'], 0.75) - np.quantile(df2['price'], 0.25)
lower = np.quantile(df2['price'], 0.25) - 3*iqr
upper = np.quantile(df2['price'], 0.75) + 3*iqr

out_df = np.where((df2['price'] < lower) | (df2['price'] > upper))
df3 = df2.reset_index().drop(out_df[0])
len(df3)
```

```python
plt.subplot(1, 2, 1)
plt.boxplot(df2['price'])
plt.subplot(1, 2, 2)
plt.boxplot(df3['price'])
plt.tight_layout(w_pad = 0.95)
#different axes
#potentially extract with PyTorch
```

```python
df3.shape
```

Split (validation later ???)

```python
X = df3.drop(columns = ['price'])
y = df3['price']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y ,random_state=23, test_size=0.2)
```

Scaling y

```python
from sklearn.preprocessing import MinMaxScaler

y_scaler = MinMaxScaler()
y = y_scaler.fit_transform(df3['price'].values.reshape(-1, 1))
```

Scaling and OneHotEncoding via Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

catcol = ['instant_bookable']
numcols = X.select_dtypes(include=['float64', 'int64']).columns

cat_transformer = Pipeline(steps = [
    ('oh_encoder', OneHotEncoder(handle_unknown='ignore'))])

num_transformer = Pipeline(steps=[
    ('mm_scaler', MinMaxScaler())])

prep = ColumnTransformer(
    transformers = [
        ('num', num_transformer, numcols),
        ('cat', cat_transformer, catcol)])

X_train_prepped = prep.fit_transform(X_train)
X_test_prepped = prep.transform(X_test)
```

To tensors

```python
X_train_tensor = torch.tensor(X_train_prepped, dtype = torch.float32)
X_test_tensor = torch.tensor(X_test_prepped, dtype = torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype = torch.float32).reshape(-1, 1)
y_test_tensor = torch.tensor(y_test.values, dtype = torch.float32).reshape(-1, 1)
```

Define class of NN

```python
#num of features is input dimensions
in_dims = X_train_tensor.shape[1]

class NNModel(nn.Module):
  def __init__(self, in_dims):
    super(NNModel, self).__init__()
    #usually around 2/3 of input size plus output layers IN USE: 3 layers
    #start small
    #or mean of input and output layers
    self.fc1 = nn.Linear(in_dims, 64)
    self.fc2 = nn.Linear(64, 32)
    self.fc3 = nn.Linear(32, 1)
    #Linear layers: non-linear with activation funcs

  #forward pass, leaky ReLU used
  def forward(self, x):
    x = torch.nn.functional.leaky_relu(self.fc1(x))
    x = torch.nn.functional.leaky_relu(self.fc2(x))
    #what is this line for ???
    x = self.fc3(x)
    return x

#instantiate nn
#HOW DOES THIS LINE WORK
nn_model = NNModel(in_dims)
```

Counting parameters (create function)

PyTorch

```python
loss_func = nn.MSELoss()
#vary learning rate good start for Adam is 0.001
opt = optim.Adam(nn_model.parameters(), lr = 0.001)
#parameters refers to weights and biases
```

```python
train_losses = []
epochs = 500   #small dataset reccommends 10- 50, IN USE: 25 (better with 250)

for epoch in range(epochs):
  #avoids accumulation from back prop (check reasoning)
  #no overlap from other data (again, check)
  opt.zero_grad()
  #pass training data
  outputs = nn_model(X_train_tensor)
  #calculate loss, then backward propagate to ...
  loss = loss_func(outputs, y_train_tensor)
  #what goes on
  loss.backward()
  opt.step()
  #why item()
  train_losses.append(loss.item())
  #check every 25 epochs
  if (epoch ) % 25 == 0:
    print(f"Epoch num: {epoch}, Loss:  {loss.item()}")
```

Plot training process

```python
plt.plot(train_losses)
plt.show()
```

```python
plt.title("Training Curve")
plt.xlabel("Epochs")
plt.plot(train_losses, color = 'firebrick')
plt.ylabel("Mean Sqaured Error")
plt.grid(alpha=0.75)
plt.show()
```

Evaluation

```python
with torch.no_grad():
  predicted_values = nn_model(X_test_tensor)
  test_loss = loss_func(predicted_values, y_test_tensor)
  print(f'Test Loss: {test_loss.item():.4f}')
```
