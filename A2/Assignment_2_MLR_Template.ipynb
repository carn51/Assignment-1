Loading data

```r
data = read.csv("https://raw.githubusercontent.com/uqglmn/sm/refs/heads/main/A2/A2_01.csv")
head(data)
```

```r
y = data$y
x1 = data$x1
x2 = data$x2
x3 = data$x3
x4 = data$x4
x5 = data$x5
x6 = data$x6
```

Scatter Plot

```r
pairs(y ~ x1 + x2 + x3 + x4 + x5 + x6, data = data,
      col='red')
```

Model fitting

```r
model = lm(formula = y ~ x1 + x2 + x3 + x4 + x5 + x6, data = data)
summary(model)
```

#### 1.1. Residual analysis

```r
options(repr.plot.width=16, repr.plot.height=4)
par(mfrow = c(1, 4), mar=c(4,4,2,1), cex=1.3)

res = residuals(model)
std_res = rstandard(model)
y_hat = fitted(model)

plot(model)
hist(std_res)
plot(y_hat, std_res)
plot(x1, y, main = "x1")
plot(x2, y, main = "x2")
plot(x3, y, main = "x3")
plot(x4, y, main = "x4")
plot(x5, y, main = "x5")
plot(x6, y, main = "x6")
```
