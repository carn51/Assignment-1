Architecture: better to overfit then regularize

```python
input_neuron_num = 12
class_num_ouput = 5    #softmax
hidden_layers_num = 2
input_layer_1 = 16
input_layer_2 = 8
output_layer_num = 5
num_epochs = 40
learning_rate = 0.1    #for now then scale by 10
#half the number that makes it diverge
gradient_descent_param = .9
#sigmoid activation
#glorot initialisation
```

Building: He Uniform, initialise all a once

```python
limit1 = np.sqrt(6/input_neuron_num)
W1 = np.random.uniform(low=-limit1, high=limit1, size=(input_neuron_num, input_layer_1))
b1 = np.zeros((1, input_layer_1))

limit2 = np.sqrt(6 / input_layer_1)
W2 = np.random.uniform(low=-limit2, high=limit2, size=(input_layer_1, input_layer_2))
b2 = np.zeros((1, input_layer_2))

limit3 = np.sqrt(6 / input_layer_2)
W3 = np.random.uniform(low=-limit3, high=limit3, size=(input_layer_2, output_layer_num))
b3 = np.zeros((1, output_layer_num))
```

Activation Function (ReLU currently)

```python
def relu(data):
  return np.maximum(0, data)
```

Dot product: first forward pass

```python
print(X_train2.shape)
print(X_train2[0])
```

```python
Z1 = (X_train2 @ W1) + b1
H1 = relu(Z1)

Z2 = (H1 @ W2) + b2
H2 = relu(Z2)

Z3 = (H2 @ W3)+ b3
#this is new output
exp_scores = np.exp(Z3)
probabilities = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)
```

Try cat-cross entropy then hinge loss. both have their advantages

```python
from sklearn.metrics import log_loss
#same as categorical cross entropy loss

ce_loss = log_loss(y_train_encoded, probabilities)
print(ce_loss)
```
