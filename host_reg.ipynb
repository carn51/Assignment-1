# Load Data

```python
## Load first chosen dataset into a Pandas (or PySpark/Dask) DataFrame
data_syd = pd.read_csv('/content/listings_syd.csv.gz')
df_syd = pd.DataFrame(data_syd)
df_syd['city'] = 'Sydney'

data_mel = pd.read_csv('/content/listings_mel.csv.gz')
df_mel = pd.DataFrame(data_mel)
df_mel['city'] = 'Melbourne'

data_bri = pd.read_csv('/content/listings_bri.csv.gz')
df_bri = pd.DataFrame(data_bri)
df_bri['city'] = 'Brisbane'
```

# Clean Data

Concatenate and select relevant columns

```python
## A couple of lines to orientate oneself wrt the data, columns and dtypes.

df0 = pd.concat([df_syd, df_bri, df_mel], axis=0)
#df0.info()

#hdf = host df
hdf = df0[['host_since',
           'host_name',
           'host_about',
           'host_acceptance_rate',
           'host_has_profile_pic',
           'host_total_listings_count',
           'host_is_superhost',
           'host_verifications',
           'description',
           'city']]

hdf.head()
```

String Cleaning and Dropping NANs

```python
## Do any appropriate cleaning

hdf = hdf.dropna()

#cleaning host_amount for sentiment analysis
import re

def clean_func(string):
  #ensure text is in fact strings
  text = str(string)
  #remove tags
  cleaner_text = re.sub('<[^<]+?>', '', text)
  full_clean_text = re.sub(r'[^a-zA-Z0-9\s]', '', cleaner_text)
  return full_clean_text

hdf['clean_abt'] = hdf['host_about'].apply(clean_func)
hdf['clean_desc'] = hdf['description'].apply(clean_func)

hdf = hdf.drop(columns = ['description', 'host_about'])

hdf.head(3)
```

Experience of host

```python
import datetime

#experience of hosts
current = datetime.datetime.now()
hdf['host_since'] = pd.to_datetime(hdf['host_since'])
hdf['host_exp'] = ((current - hdf['host_since']) / pd.Timedelta(days= 365.25))

#host accceptance rate float
hdf['host_acceptance_rate'] = hdf['host_acceptance_rate'].str.replace('%', '').astype('float')
hdf = hdf.drop(columns = ['host_since', 'host_verifications'])

hdf.head(3)
```

```python
#import back in
hdf2 = pd.read_csv('/content/host_df')


#names can be dropped fro our purposes moving forward
hdf2 = hdf2.drop(columns = ['host_name'])
hdf2.info()
```

```python
hdf2['host_has_profile_pic'] = hdf2['host_has_profile_pic'].astype('category')
hdf2['host_is_superhost'] = hdf2['host_is_superhost'].astype('category')

from sklearn.preprocessing import OneHotEncoder

oh_pic = OneHotEncoder(handle_unknown='ignore', sparse_output = False)
onehot_pic = oh_pic.fit_transform(hdf2[['host_has_profile_pic']])
#to df
columns_pic=oh_pic.get_feature_names_out(['host_has_profile_pic'])
onehot_pic_df = pd.DataFrame(onehot_pic, columns = columns_pic)

oh_sh = OneHotEncoder(handle_unknown='ignore', sparse_output = False)
onehot_superhost = oh_sh.fit_transform(hdf2[['host_is_superhost']])
columns_sh=oh_sh.get_feature_names_out(['host_is_superhost'])
onehot_sh_df = pd.DataFrame(onehot_superhost, columns= columns_sh)

oh_city = OneHotEncoder(handle_unknown='ignore', sparse_output = False)
onehot_city = oh_city.fit_transform(hdf2[['city']])
columns_city=oh_city.get_feature_names_out(['city'])
onehot_city_df = pd.DataFrame(onehot_city, columns=columns_city)

#third version
hdf3 = pd.concat([hdf2, onehot_pic_df, onehot_sh_df, onehot_city_df], axis=1).drop(columns = ['host_has_profile_pic', 'host_is_superhost', 'city'])
hdf3.head(3)
```

Final Prep: Sentiment analysis

```python
from textblob import TextBlob

sentiment_list = []
polarity_list = []

def sentiment_func(text):
  analysis = TextBlob(text)
  return analysis.sentiment.polarity

#impute with empty strimngs, drop later if needed
hdf3['clean_abt'] = hdf3['clean_abt'].fillna('')

hdf3['sentiment_desc'] = hdf3['clean_desc'].apply(sentiment_func)
hdf3['sentiment_abt'] = hdf3['clean_abt'].apply(sentiment_func)

hdf3 = hdf3.drop(columns = ['clean_abt', 'clean_desc'])
hdf3.head(3)
```

Split

```python
from sklearn.model_selection import train_test_split

y = hdf3['host_total_listings_count']
X = hdf3.drop(columns=['host_total_listings_count'])

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    random_state=23
)
```

Start with model: Instatntiate, Train, Predict

(scaling may be needed, if yes when)

```python
from sklearn.ensemble import RandomForestRegressor

rf_reg = RandomForestRegressor(
    n_estimators=100,
    random_state=42,
    oob_score=True)

rf_reg.fit(X_train, y_train)
y_pred = rf_reg.predict(X_test)
```

Evaluate

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)
```
